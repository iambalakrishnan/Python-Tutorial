```python
from threading import *
from time import sleep
class Hello(Thread):
    def run(self):
        for i in range(10):
            print("Hello")
            sleep(1)
```

```python
class Hi(Thread):
    def run(self):
        for i in range(10):
            print("Hi")
            sleep(1)
```

```python
t1=Hello()
t2=Hi()
```

```python
t1.start()
sleep(0.2)
t2.start()
```

```python
t1.start()
sleep(0.2)
t2.start()
print("bye")
```

```python
t1.start()
sleep(0.2)
t2.start()

t1.join()
t2.join()
print("bye")
```
